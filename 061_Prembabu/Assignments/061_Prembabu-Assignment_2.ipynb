## Split the data into dependent and independent variables


```python
x=ds.iloc[:,3:13].values
print(x.shape)
y=ds.iloc[:,13:14].values
print(y.shape)
```

## Finding and Replacing Outliers

```python
ds.skew()
```

```python
sns.boxplot(ds["Age"])
```

```python
q0 = ds["Age"].describe()["25%"]
q1 = ds["Age"].describe()["75%"]
iqr=q1-q0
```

```python
lb = q0 -(1.5*iqr)
ub = q1 + (1.5*iqr)
```

```python
ds[ds["Age"]<lb]
```

```python
ds[ds["Age"]>ub]
```

```python
outlier_list = list(ds[ds["Age"] > ub]["Age"])
```

```python
print(outlier_list)
```

```python
outlier_dict = {}.fromkeys(outlier_list,ub)
```

```python
print(outlier_dict)
```

```python
ds["Age"] = ds["Age"].replace(outlier_dict)
sns.boxplot(ds["Age"])
```

```python
ds[ds["Age"]>ub]
```

## Check for Categorical columns and perform encoding

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer([('oh',OneHotEncoder(),[1,2])],remainder='passthrough')
x=ct.fit_transform(x)
print(x.shape)
```

```python
import joblib
joblib.dump(ct,"churnct.pkl")
```
